## Solve Ordinary Differential Equations (ODE)

Consider the following first order diffeential equation: \\
$m \frac{dv}{dt} + kv(t) = mg$

Here $v(t)$ is a function of $t$ and we have to find the solution $v(t)$.

Use $dsolve()$ funcion to solve differential equations. \\
**Format:** $dsolve(Eq(LHS, RHS), y(t))$ \\
Here LHS and RHS should be differential terms + normal terms.

```python
import numpy as np
from sympy import *

m, g, k = 1, 9.8, 1

t = symbols('t')

v = Function('v')

sol = dsolve(Eq(m*v(t).diff(t) + k*v(t), m*g), v(t))
sol

# It is a first order differential equation and thus we have one constant C1 to be determined by an initial condition.
```

```python
# Let's consider the initial condition v(t = 0) = 0.

print(sol.lhs)  # LHS of the solution.
print(sol.rhs)  # RHS of the solution.

Eq(sol.rhs.subs(t, 0), 0)  # Initial condition creates an equation.
```

```python
solve(Eq(sol.rhs.subs(t, 0), 0))  # Solve the initial condition applied equation and find the constants.
```

```python
C1_val = symbols('C1_val')
C1_val = solve(Eq(sol.rhs.subs(t, 0), 0))[0]  # Taking the first (or only) element of the solution list.
C1_val
```

```python
# Substitute the value of the constant and find the final solution v(t).

C1 = symbols('C1')
sol = sol.subs(C1, C1_val)

sol
```

```python
# Plot the solution v(t).

import matplotlib.pyplot as plt

plot(sol.rhs, (t, 0, 10))
```

Now consider the second order differential equation: \\
$\frac{d^2y}{dt^2} + y(t) = 0$

```python
t = symbols('t')

y = Function('y')

sol = dsolve(Eq(y(t).diff(t, t) + y(t), 0), y(t))
sol
```

```python
# Now use the initial conditions: y(0) = 0 and y'(0) = 1, where y' = dy/dt, and calculate C1 and C2.

sol.rhs.subs(t, 0)
```

```python
C1_val = solve(Eq(sol.rhs.subs(t, 0), 0))[0]
C1_val
```

```python
sol.rhs.diff(t).subs(t, 0)
```

```python
C2_val = solve(Eq(sol.rhs.diff(t).subs(t, 0), 1))[0]
C2_val
```

```python
C1, C2 = symbols('C1, C2')
sol = sol.subs({C1: C1_val, C2: C2_val})
sol
```

```python
# Plot the solution.

plot(sol.rhs, (t, 0, 10))
```
