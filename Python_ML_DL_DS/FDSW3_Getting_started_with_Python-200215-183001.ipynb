```python
import sys

weight = input('Enter your weight (in Kgs): ')
unit_of_height = input('What is your preferred unit of height (Type F for feet and M for meter)? ')

if unit_of_height == 'F':
  feet = input('Enter your height in feet: ')
  inch = input('Now enter inches: ')
  height = eval(feet)*0.3048 + eval(inch)*0.0254
elif unit_of_height == 'M':
  meter = input('Enter your height in meter: ')
  height = eval(meter)
else:
  print('Type F or M in unit of height')
  sys.exit()

BMI = eval(weight)/(height**2)
print('Your BMI value is: ', BMI)

if BMI < 18.5:
  print('You are UNDERWEIGHT!')
elif BMI >= 18.5 and BMI < 25:
  print('You have NORMAL weight. Very good!')
elif BMI >= 25 and BMI < 30:
  print('You are OVERWEIGHT!')
elif BMI >= 30:
  print('You are VERY-OVERWEIGHT!!!')
```

```python
# Problem-1

def factorial(n):
  facto = 1
  for i in range(1, n+1):
    facto = facto * i
  return facto

print('The factorial using iterative function: ', factorial(5))

# Problem-2

def factorial_recursive(n):
  if n > 1:
    return n * factorial_recursive(n-1)
  else:
    return 1
  print(facto)

print('The factorial using recursive function: ', factorial_recursive(6))

# Problem-3

def function_nth_term(x, n):
  nth_term = x**n/factorial(n)
  return nth_term

print('The nth term for a given x and n:', function_nth_term(2.3, 5))

# Problem-4

def function_sum_upto_Nth_term(x, N):
  sum = 1
  for n in range(1, N+1):
    sum = sum + function_nth_term(x, n)
  return sum

print('The sum upto the nth term: ', function_sum_upto_Nth_term(1, 10))

# Problem-5

def function_difference_betn_two_sums(x, N, epsilon):
  delta_sum = (function_sum_upto_Nth_term(x, N) - function_sum_upto_Nth_term(x, N-1))
  while delta_sum < epsilon:
    print('Convergence achieved')
    print('The difference between the sum of Nth term and (N-1)th term: ', delta_sum)
    break

function_difference_betn_two_sums(1, 10, 0.000001)

# Problem-6
p = 2.1
q = 1.3

v1 = function_sum_upto_Nth_term(p, 100) * function_sum_upto_Nth_term(q, 100)

v2 = function_sum_upto_Nth_term(p + q, 100)

print('The values are v1, v2, v1-v2: ', v1, v2, v1 - v2)
```

```python
# Asissgnment-2: Problem-5 (solution by Pratyush Kumar)

def function_difference_betn_two_sums_Pratyush_Kumar(x, epsilon):
  sum = 1
  var = epsilon  # this tells to go into the while loop
  n = 1
  while var >= epsilon:
    var = function_nth_term(x, n)
    sum = sum + var
    n = n + 1
  return sum

print(function_difference_betn_two_sums_Pratyush_Kumar(2, 0.01))

# Another way:

def function_difference_betn_two_sums_Pratyush_Kumar_II(x, epsilon):
  sum = 1
  n = 1
  while True:      # no condition on var previously. Always enter into the while loop. 
    var = function_nth_term(x, n)
    sum = sum + var
    n = n + 1
    if var < epsilon:
      break
  return sum

print(function_difference_betn_two_sums_Pratyush_Kumar_II(2, 0.01))
```
