## Integration of Normal distribution

```python
import numpy as np
from scipy.integrate import quad

mu = 7.8
sigma = 0.6

f = lambda x: (1 / (np.sqrt(2*np.pi) * sigma)) * np.exp(-((x-mu)**2) / (2*sigma**2))

quad(f, 6.6, 9.0)
```

## Median of normal distribution

Find the median of the random variable Y = max(0, X), where X is also a random variable following normal distribution of zero mean and unit variance.

```python
import numpy as np
import matplotlib.pyplot as plt

N = np.random.normal(0, 1, 100000)
# print(N)

plt.hist(N, bins = 30)
plt.show()
```

```python
# N is basically X here.

Y = []
for i in N:
    if i < 0:
        Y.append(0)
    else:
        Y.append(i)

# print(Y)
# print(N[0:10])

plt.hist(Y, bins = 30)
plt.show()

Y = np.array(Y)
print(np.median(N), np.median(Y))

# Median of both X and Y are 0.
```
