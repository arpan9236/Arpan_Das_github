```python
from os import replace
import numpy as np
import sys

total_seats = np.arange(1, 73)
print('The array of all the seats: ', total_seats)

compartments = np.split(total_seats, 9)
print('The array of each compartment: ', compartments)

booked_seats = []

for i in range(40):
    x = np.random.randint(1, 73)
    if x not in booked_seats:
        booked_seats.append(x)

booked_seats = np.random.choice(range(1, 73), size = 40, replace = False)

print('The array of already booked seats: ', booked_seats)
print('Total number of booked seats = ', len(booked_seats))

# Number of seats I want to book
n = 6

A = []

j = 0
for i in range(1, 73):
    if i not in booked_seats:
        A.append(i)

print('The seat numbers which are not booked yet: ', A)
print('The number of non-booked seats = ', len(A))

B = []

for i in A:
    for j in range(9):
        if i in compartments[j]:
            B.append(j)

print('The array of compartment indices of all the non-booked seats: ', B)
print('Compartment indices and the number of non-booked seats in each compartment: ', np.unique(B, return_counts=True))

compartment_index, count = np.unique(B, return_counts=True)

for i in compartment_index:
    if count[i] >= n:
        print('The compartment index in which your all the', n ,'seats are booked: ', compartment_index[i])
        # print(np.sum(count[:i]))
        print('Your seat numbers are = ', A[np.sum(count[:i]):np.sum(count[:i]) + n])
        break
else:
    print('The seats are not available in the same comparment!!')
```

# 2 Dice Rolling

```python
# Identify all the outcomes when 2 dice are rolling.

A = []   # List of all outcomes.

for i in range(1, 7):
    for j in range(1, 7):
        A.append((i, j))

print('The outcomes forming the sample space: ', A)
print('The total number of outcomes =', len(A))

# Selecting only the outcomes which have c
c = 6

B = []   # List of outcomes which have c.

for i, j in A:
    # print(i, j)
    if (i == c) or (j == c):
        # print(i, j)
        B.append((i, j))

print('The outcomes which have', c, '=', B)
print('The number of outcomes which have', c, '=', len(B))

print('The probability of getting', c, 'in 2 dice rolling =', len(B) / len(A))
```

# 3 Dice Rolling

```python
# Identify all the outcomes when 3 dice are rolling.

A = []   # List of all outcomes.

for i in range(1, 7):
    for j in range(1, 7):
        for k in range(1, 7):
            A.append((i, j, k))

print('The outcomes forming the sample space: ', A)
print('The total number of outcomes =', len(A))

# Selecting only the outcomes which have c
c = 6

B = []   # List of outcomes which have c.

for i, j, k in A:
    # print(i, j, k)
    if (i == c) or (j == c) or (k == c):
        # print(i, j, k)
        B.append((i, j, k))

print('The outcomes which have', c, '=', B)
print('The number of outcomes which have', c, '=', len(B))

print('The probability of getting', c, 'in 2 dice rolling =', len(B) / len(A))
```

## Sample statistic as discrete RV

### Proof of $E[\bar{X}] = \mu$ for discrete RV or discrete sample statistic

$\bar{X} = $ A Random Variable (RV). We take sample statistic as a RV. Here, sample mean is a RV.

We want to know the 'mean income' of population consists of all the people of India. $\mu = $ population mean. $N$ = Population size.

We take many many samples. Each sample size = $n$. So, total number of possible samples, $k =$ $N \choose n$. In real world, we do not take all the samples as $k$ becomes very very large. We take some random samples out of total $k$ samples. So $k$ is basically number of samples considered in our analysis.

We first calculate mean income of all the samples. Each sample gives one value of sample mean $\bar{X}$. So, $\bar{X}$ is the sample mean and it is a RV which takes many values coming from all the $k$ samples. We also assign probability to all the values that the RV $\bar{X}$ can take and thus we get the distribution of sample statistic $\bar{X}$.

Mean of $\bar{X}$ of all the $k$ samples $\approx E[\bar{X}]$ for large $k$. So, we will prove that for large $k$, mean of $\bar{X}$ of all the $k$ samples $\approx \mu$, the population mean.

In this case sample mean income $\bar{X}$ is a continious RV.

In the follwoing first we will do the exercise of discrete RV in the case of dice rolling. And then we will do the same exercise for the continious RV.

```python
# Dice rolling.

import random
from math import comb
import numpy as np

# Population size
N = 1000

# Sample size
n = 2

# Combination or number of all possible samples
k = comb(N, n)
print('Number of all possible samples = ', k)

# Create list of outcomes
outcomes = []
for i in range(1, 7):
    outcomes.append(i)

print('The outcomes of the dice rolling experiment: ', outcomes)

# Dice rolls N times. Outcomes are generated randomly. This makes the population.
A = []
for i in range(N):
    A.append(random.choice(outcomes))

# Calculation population mean \mu
# print(A)
print('The population mean = ', sum(A) / len(A))

# Create all the samples in the population space Omega
Omega = []
for i in range(k):
    sample = []
    for j in range(n):
        sample_elements = random.choice(outcomes)
        sample.append(sample_elements)
    Omega.append(tuple(sample))

# print(Omega)


# Calculate sample mean for all the samples.
X_bar = []
for i in Omega:
    X_bar.append(np.mean(i))

# print(X_bar)
print('Sample mean or the mean of all the samples = ', np.mean(X_bar))


# In real world problems, we sometimes do not take all possible k samples as it becomes very very large. We take same random samples out of all k samples.

# We approximate that as the number of samples, we consider, increases, the mean of all the mean_income values of the samples considered become closer to
# the expected value of the mean_income of the samples considered. This expected value of the mean_income of all possible samples is equal to the
# population mean.

# We see two observations: (i) even the individual value of each sample is different from population mean = 3.5, if you take
# many many samples and take the mean of all the samples, the value will be very close to the population mean. (ii) if the sample size n increases,
# the mean of all the sample mean closes to population mean rapidly. The deviation from mean or the spread of mean_income of all the samples decreases
# as the sample size n increases. That is, as the sample size n increases, variance of the all the sample mean becomes smaller. The values of
# sample mean become narrower about the population mean.
```

### Relation between Var($\bar{X}$) and Sample size $n$



$N = $ Population size \\
$n = $ Sample size  \\
$k = $ Number of samples $=$ $N \choose n$

So, roll the dice $n$ times and note the outcomes. This set of $n$ outcomes make a sample. So, each sample size is $n$. Thus you make total $k = $ $N \choose n$ total possible samples.

For each sample calculate the mean of $n$ outcomes. This is the sample mean. Define sample mean $\bar{X}$ as a RV. So, $\bar{X}$ is a discrete RV and which takes different values coming from $k$ different samples. Then calculate the variance (which is a measure of spread) of $\bar{X}$.

Change the sample size $n$ and each time calculate the variance of $\bar{X}$. See the change of Var($\bar{X}$) with the sample size $n$. Find a functional relationship by fitting different curve to Var($\bar{X}$) vs $n$ plot.

```python
# Create all the samples in the probability space Omega

# Function for calculation of variance for each sample size n
def calc_variance(n):
    k = 1000
    Omega = []
    for i in range(k):
        sample = []
        for j in range(n):
            sample_elements = random.choice(outcomes)
            sample.append(sample_elements)
        Omega.append(tuple(sample))
    # return Omega

    X_bar = []
    for i in Omega:
        # print(i)
        X_bar.append(np.mean(i))
    return np.var(X_bar)

# print(calc_variance(3))


# Variance of X_bar for different sample size n
n = []
Var_of_X_bar = []
for i in range(2, 100):
    n.append(i)
    Var_of_X_bar.append(calc_variance(i))

n = np.array(n)
print(n)
Var_of_X_bar = np.array(Var_of_X_bar)
print(Var_of_X_bar)


# Plot
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the model function or fit function for curve fitting
def fit_func(n, a, b):
    return a * np.exp(-b * n)

# Parameters of the curve fit
param, param_cov = curve_fit(fit_func, n, Var_of_X_bar)

# The fitted curve
fitted_curve = param[0] * np.exp(-param[1] * n)
print(param)

# Plotting the actual data and the fitted curve
plt.scatter(n, Var_of_X_bar)
plt.plot(n, fitted_curve, color = 'red')
plt.title('Var(X_bar) vs sample size n')
plt.xlabel('Sample size n')
plt.ylabel('Var(X_bar)')
plt.show()
```

```python
# Plot
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the model function or fit function for curve fitting
def fit_func(n, c):
    return (c / n)

# Parameters of the curve fit
param, param_cov = curve_fit(fit_func, n, Var_of_X_bar)

# The fitted curve
fitted_curve = param[0] / n
print(param)

# Plotting the actual data and the fitted curve
plt.scatter(n, Var_of_X_bar)
plt.plot(n, fitted_curve, color = 'red')
plt.title('Var(X_bar) vs sample size n')
plt.xlabel('Sample size n')
plt.ylabel('Var(X_bar)')
plt.show()
```

So, we see that the plot of Var($\bar{X}$) vs sample size $n$ follows $c/n$ decay instead of exponential $a e^{-bn}$ decay (where $a, b, c$ are constants or parameters of the curve fitting).

### Var($\bar{X}$) vs $\mu$: Variance of sample mean vs population mean

Take continious RV $\bar{X} = $ mean of 10 numbers from normal distribution $N(\mu, \sigma)$. Sample size $n = 10$.

1. Compute Var($\bar{X}$) over $k = 1000$ samples.

2. Change population mean $\mu$ and recompute Var($\bar{X}$). Plot Var($\bar{X}$) vs $\mu$.

3. Change population standard deviation $\sigma$ and recompute Var($\bar{X}$). Plot Var($\bar{X}$) vs $\sigma$.

```python
# General function to calculate Var(X_bar)

import random
import numpy as np
import matplotlib.pyplot as plt

# Define a general function with general value of N, n, k, mu and sigma for calculation of variance
def calc_variance(N, n, k, mu, sigma):
    outcomes = np.random.normal(mu, sigma, N)

    Omega = []
    for i in range(k):
        sample = []
        for j in range(n):
            sample_elements = random.choice(outcomes)
            sample.append(sample_elements)
        Omega.append(tuple(sample))
    # return Omega

    X_bar = []
    for i in Omega:
        # print(i)
        X_bar.append(np.mean(i))
    return np.var(X_bar)

# Check the function
print(calc_variance(N = 1000, n = 10, k = 1000, mu = 0, sigma = 1))
```

```python
# Var(X_bar) vs mu

mu = []
Var_of_X_bar = []
for i in np.arange(-5, 5, 0.1):
    mu.append(i)
    Var_of_X_bar.append(calc_variance(N = 1000, n = 10, k = 1000, mu = i, sigma = 1))

mu = np.array(mu)
print('The values of population mean = ', mu)
Var_of_X_bar = np.array(Var_of_X_bar)
print('The values of variance of sample mean = ', Var_of_X_bar)


# Plot
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the model function or fit function for curve fitting
def fit_func(mu, c):
    return (c / mu)

# Parameters of the curve fit
param, param_cov = curve_fit(fit_func, mu, Var_of_X_bar)

# The fitted curve
fitted_curve = param[0] / mu
print(param)

# Plotting the actual data and the fitted curve
plt.plot(mu, Var_of_X_bar)
# plt.plot(mu, fitted_curve, color = 'red')
plt.title('Var(X_bar) vs population mean mu')
plt.xlabel('Population mean mu')
plt.ylabel('Var(X_bar)')
plt.show()
```

### Var($\bar{X}$) vs $\sigma$: Variance of sample mean vs population S.D.

```python
# Var(X_bar) vs sigma

sigma = []
Var_of_X_bar = []
for i in np.arange(0, 10, 0.2):
    sigma.append(i)
    Var_of_X_bar.append(calc_variance(N = 1000, n = 10, k = 1000, mu = 0, sigma = i))

sigma = np.array(sigma)
print('The values of population S.D. = ', sigma)
Var_of_X_bar = np.array(Var_of_X_bar)
print('The values of variance of sample mean = ', Var_of_X_bar)


# Plot
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the model function or fit function for curve fitting
def fit_func(sigma, a):
    return (a * sigma**2)

# Parameters of the curve fit
param, param_cov = curve_fit(fit_func, sigma, Var_of_X_bar)

# The fitted curve
fitted_curve = param[0] * sigma**2
print(param)

# Plotting the actual data and the fitted curve
plt.scatter(sigma, Var_of_X_bar)
plt.plot(sigma, fitted_curve, color = 'red')
plt.title('Var(X_bar) vs population S.D. sigam')
plt.xlabel('Population S.D. sigma')
plt.ylabel('Var(X_bar)')
plt.show()
```

**So we have 3 hypothesis:**

1. Var($\bar{X}$) is inversly proportional to the sample size $n$: Var($\bar{X}$) $= \frac{c}{n}$.

2. Var($\bar{X}$) is independent of the population mean $\mu$.

3. Var($\bar{X}$) is parabolically increasing with population standard deviation $\sigma$: Var($\bar{X}$) $= a \sigma^2$.  

Combining these 3 hypothesis, we get the genral formula of Var($\bar{X}$) as following,

Var($\bar{X}$) $= \frac{\sigma^2}{n}$.

So, we finally got two relations,

(i) $E[\bar{X}] = \mu$

(ii) Var($\bar{X}) = \frac{\sigma^2}{n}$  

where, $\bar{X} = $ Sample mean. Sample statistic (here, mean) is a RV. \\
$\mu = $ Population mean \\
$\sigma = $ Population standard deviation \\
$n = $ Sample size.

## Use of CLT in Data Science

**Statement of the Central Limit Theorem (CLT):**

If $X_1, X_2, ..., X_n$ are random samples from a population with mean $\mu$ and standard deviation (S.D.) $\sigma$, then the sum

$X^+ = X_1 + X_2 + ... + X_n$

will converge to normal distribution with mean $n \mu$ and S.D. $\sigma \sqrt{n}$, written as $N(n \mu, \sigma \sqrt{n})$ in the limiting condition of sample size $n \to \infty$.

Clearly, the **sample mean** $\bar{X} = \frac{X^+}{n}$, \\
where $X^+ = $ **sample sum** and $n = $ sample size.  

So, $E[\bar{X}] = E[\frac{X^+}{n}] = \frac{1}{n} E[X^+] = \frac{1}{n} n \mu = \mu$,

and $Var(\bar{X}) = Var(\frac{X^+}{n}) = \frac{1}{n^2} Var(X^+) = \frac{1}{n^2} \sigma^2 n = \frac{\sigma^2}{n}$.

So, $S.D.(X^+) = \sqrt{\frac{\sigma^2}{n}} = \frac{\sigma}{\sqrt{n}}$.

Therefore, in the limit of sample size $n \to \infty$,

(i) distribution of **sample sum** $X^+$ converges to $N(n \mu, \sigma \sqrt{n})$,

(ii) distribution of **sample mean** $\bar{X} = \frac{X^+}{n}$ converges to $N(\mu, \frac{\sigma}{\sqrt{n}})$.

### Compute liklihood of sample mean

We have to take many many samples and compute $X^+$ or $\bar{X}$. $X^+$ may not be the same for all the samples i.e., $X^+$ can take many numerical values (discrete or continious depending on the problem) and we have to assign probability to each value that the RV $X^+$ can take. Thus we get probability distribution (PMF for discrete RVs and PDF for continious RVs) of $X^+$ or $\bar{X}$.

**Likelihood of sample mean** means if we know population mean $\mu$ and population S.D. $\sigma$ and have a measurement of sample sum $X^+$ or sample mean $\bar{X}$, we can compute probability of getting $X^+ > a$ or $X^+ < a$, say some number.

The probability $Pr(X^+ > a)$ or $Pr(X^+ < a)$ can not be said only from the knowledge of $\mu$, $\sigma$ or $E[X^+]$, $Var(X^+)$. We need to have the knowledge of probability distribution (PMF for discrete RVs and PDF for continious RVs) of $X^+$ or $\bar{X}$.

Let's do an exercise to understand the use of CLT in Data Science.

We have the population consists of millions of dice rolling. Take a random samples of 10 dice rolling and compute the sum of 10 numbers which is the sample sum $X^+ = X_1 + X_2 + ... + X_{10}$. What is the probability of getting $X^+ > 40$ i.e., compute $Pr(X^+ > 40)$. Here the sample size $n = 10$.

Each of the dice rolling is independent and they form uniform distribution of outcomes 1, 2, 3, 4, 5, 6. So, each $X_i$ can take a value out of this 6 discrete values. So, the population has uniform distribution $U(1, 6)$.

CLT tells, in the limit of $n \to \infty$, distribution of $X^+$ converges to normal distribution $N(n \mu, \sigma \sqrt{n})$. Though $n = 10$ only here, but we will check.

Here population mean, $\mu = \sum_{i=1}^6 x_i p(x_i) = 1 \times \frac{1}{6} + 2 \times \frac{1}{6} + 3 \times \frac{1}{6} + 4 \times \frac{1}{6} + 5 \times \frac{1}{6} + 6 \times \frac{1}{6} = \frac{1}{6} \times 21 = 3.5$,

and population S.D. $\sigma = \sqrt{\frac{1}{6} \sum_{i=1}^6 (x_i - \mu)^2} = \sqrt{\frac{1}{6} [(1 - 3.5)^2 + (2 - 3.5)^2 + (3 - 3.5)^2 + (4 - 3.5)^2 + (5 - 3.5)^2 + (6 - 3.5)^2]} = 1.708$ .

Therefore, using CLT, distribution of $X^+$ converges to the normal distribution $N(n \mu, \sigma \sqrt{n}) = N(35, 5.401)$,

where $E[X^+] = n \mu = 10 \times 3.5 = 35$ and $S.D.(X^+) = \sigma \sqrt{n} = 1.708 \times \sqrt{10} = 5.401$.

Now finding the probability $Pr(X^+ > 40)$ means computing area under the normal distribution $N(35, 5.401)$, in the range [40, $+\infty$]. We will do this using the following python code.

```python
import numpy as np
from scipy.integrate import quad

# Define normal distribution of mean = 35 and S.D. = 5.401.
# N = np.random.normal(35, 5.401, size = 5000)

mu = 35
sigma = 5.401

f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Perform the integration in the range [40, +infinity]. We take a very big number as +infinity.
print(quad(f, 40, 10000))

# So, area under the normal distribution N(35, 5.401) in the range [40, +infinity] is 0.177. Hence, Pr(X^+ > 40) = 0.177.
```

Computing area under the $N(35, 5.401)$ is equivalent to computing area under the standard normal distribution $N(0, 1)$, using Z-score. In this case, instead of computing $Pr(X^+ > 40)$ we have to compute $Pr(\frac{X - \mu}{\sigma} > \frac{40 - 35}{5.401}) = Pr(Z > \frac{40 - 35}{5.401}) = Pr(Z > 0.926)$.

So, computing area under the $N(35, 5.401)$ in the range $[40, +\infty]$ is equaivalent to computing area under the standard normal distribution $N(0, 1)$ in the range $[0.926, +\infty]$.

```python
import numpy as np
from scipy.integrate import quad

# Define normal distribution of mean = 0 and S.D. = 1

mu = 0
sigma = 1

f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Perform the integration in the range [40, +infinity]. We take a very big number as +infinity.
print(quad(f, 0.926, 1000))

# So, area under the standard normal distribution N(0, 1) in the range [0.926, +infinity] is 0.177. Hence, Pr(X^+ > 40) = Pr(Z > 0.926) = 0.177.
```

It can be shown that the exact answer (w/o using CLT) of $Pr(X^+ > 40)$ from the PMF is **0.157**.

Clearly, CLT overestimates by 0.02. We have discussed the reason of this overestimation, known as **continuity correction**, in great detail in the notebook. Now we will play with the integration range to make contnuity correction and avoid over/under estimation.

So, we apply the continuity correction when we approximate PMF of a discrete RV to a normal distribution $N$ and computea area under the $N$, using CLT.

```python
import numpy as np
from scipy.integrate import quad

# Define normal distribution of mean = 35 and S.D. = 5.401.
# N = np.random.normal(35, 5.401, size = 5000)

mu = 35
sigma = 5.401

f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Perform the integration in the range [41, +infinity]. We take a very big number as +infinity.
print(quad(f, 41, 10000))

# So, area under the normal distribution N(35, 5.401) in the range [41, +infinity] is 0.133.
```

```python
print(quad(f, 40.5, 10000))

# So, area under the normal distribution N(35, 5.401) in the range [40.5, +infinity] is 0.154. This value is very close to the
# exact value of 0.157. Fantastic!!

# In this way, we play with the integration range to avoid the continuity correction.
```

Similarly, we can compute the probability that the sum of 10 dice rolling is $> 25$ and $< 36$ i.e., $Pr(25 < X^+ < 36)$.

In this case we integrate in the range $[25.5, 35.5]$ i.e., applying continuity correction on both the ends.

```python
print(quad(f, 25.5, 35.5))

# The value of Pr(25 < X^+ < 36) = 0.4976.
```

### Approximate distributions

Let's take an example. In a random group of 100 people, compute the probability mass function (PMF) of number of people having birthdays on Tuesday.

Here the RV, $X = $ number of people having birthdays on Tuesday.

We can have 0 people having birthdays on Tuesday, 1 people having birthdays on Tuesday, 2 people having birthdays on Tuesday, ..., 100 people having birthdays on Tuesday. So, the RV $X$ can take discrete values $ = 0, 1, 2, ..., 100$. We have to assign probability to all the values that the RV $X$ can take i.e., compute

$P(X = 0), P(X = 1), P(X = 2), ..., P(X = 100)$.

Thus we get the probability distribution or PMF of $X$.

Assuming no bias on any day, the probability of having birthday on Tuesday, $p = \frac{1}{7}$.

We will choose a people out of 100 people, if his/her birthday is on Tuesday, it is sucess and if not it is failure. So, this is a Bernoulli trial having two outcomes: success: if birthday is on Tuesday and failure: if birthday is not on Tuesday. We will do it for 100 people. Thus we have $n = 100$ Bernoulli trials. We have to compute probability of $k$ sucesses out of $n$ Bernoulli trials. Clearly, $k$ can take values $= 0, 1, 2, ..., n$. Thus it is a Binomial distribution:

$P (X = k) = {n \choose k} p^k (1 - p)^{(n-k)}$,

where $p = $ probability of success in a single Bernoulli trial.

Here, $n = 100, p = \frac{1}{7}$.

We will calculate $P(X = k)$ for $k = 0, 1, 2, ..., n$. This will give the PMF of the RV $X$. $P(X = k)$ is probability that the $k$ people have birthdays on Tuesday.

We know that for Binomial distribution mean $= \mu = p$ and S.D. $= \sigma = \sqrt{p (1-p)}$. With $p = \frac{1}{7}$, $\mu = 1/7$ and $\sigma = 0.35$

Therefore, the binomial distribution converges to the normal distribution $N(n \mu, \sigma \sqrt{n}) = N(14.29, 3.5)$.

```python
# PMF for Binomial distribution

import math
import numpy as np
import matplotlib.pyplot as plt

# Calculate PMF of Binomial distribution
n = 100
p = 1/7

P = []   # This list will collect the values of probabilities of all the values that the RV X can take.

for k in range(n+1):
    P.append(math.comb(n, k) * p**k * (1 - p)**(n - k))   # Prob. of Binomial distribution.

# print(P)
# print(len(P))


# Define the normal distribution N(14.29, 3.5)
mu = 14.29
sigma = 3.5

x = np.arange(0, 100, 0.1)

N = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))


# Plot
plt.plot(range(101), P, 'ro')
plt.plot(x, N, 'b-')
plt.xticks(np.arange(0, 101, step = 5))
plt.xlabel('k')
plt.ylabel('Prob. that $k$ people have birthdays on Tuesday')
plt.title('PMF of Binomial distribution and its Normal approximation $N(14.29, 3.5)$')
plt.legend(['PMF', 'Normal approximation $N(14.29, 3.5)$'])
plt.show()


# Very good approximation of Binomial distribution by N(14.29, 3.5), using CLT!!
# Now we can throw away expensive computation of Binomial distribution and can use normal distribution N(14.29, 3.5) to approximate it.
```

**Compute area under $N(14.29, 3.5)$**

$P(X = k) = P(X < k+0.5) - P(X = k - 0.5)$

Now, using CLT, we will integrate the normal distribution $N(14.29, 3.5)$, with continuity correction of 0.5, for different values of $k$, and compare with the exact result of PMF.

```python
# For k = 10

import scipy
from scipy.integrate import quad

k = 10
f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

print(quad(f, k-0.5, k+0.5))

# CLT gives 0.054. Exact answer is 0.058.
```

```python
# For k = 15

import scipy
from scipy.integrate import quad

k = 15
f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

print(quad(f, k-0.5, k+0.5))

# CLT gives 0.111. Exact answer is 0.109
```

```python
# For k = 18

import scipy
from scipy.integrate import quad

k = 18
f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

print(quad(f, k-0.5, k+0.5))

# CLT gives 0.065. Exact answer is 0.061.
```

So, we see that CLT is giving very close results to the exact answers. The difference is only at the 3rd/4th decimel places. Great!! Thanks to CLT.

Hence, we can throw away the computationally heavy exact PMF and replace it with the normal distribution using CLT, which is computationally less expensive.

**Question:** Let's say we have $X_1, X_2, ..., X_{100}$ as the given i.i.d. (independent and identically distributed random variables) along with their expected value $E[X_i] = 0.1$ and S.D. $\sigma(X_i) = 1.09$. Let $Y = X_1 + X_2 + ... + X_{100}$ is another RV. Using CLT, compute $P(1.5 \le Y \le 2.5)$.

Using CLT, the distribution of sample sum $Y$ is a normal distribution with mean $n \mu$ and standard deviation $\sigma \sqrt{n}$. In this case, sample size $n = 100$, population mean $\mu = E[X_i] = 0.1$ and population S.D. $\sigma(X_i) = 1.09$. So, mean of the normal distribution, $n \mu = 10$ and standard deviation $\sigma \sqrt{n} = 1.09$. Now $P(1.5 \le Y \le 2.5)$ is nothing but the area under the normal distribution $N(10, 10.9)$ in the range $[1.5, 2.5]$.

```python
from scipy.integrate import quad

# Mean and S.D. of normal distribution
mu = 10
sigma = 10.9

norm_dist = lambda x: (1 / (sigma * np.sqrt(2* np.pi))) * np.exp((-(x - mu)**2) / (2 * sigma**2))

print(quad(norm_dist, 1.5, 2.5))

# The answer is 0.02795.
```

A particular department inside this organization, which has 200 employees in total, gets a total of 1,440,000 INR. What is the probability that this budget is somehow insufficient for the department employees?

Using CLT, the distribution of sample sum $X^+ = X_1 + X_2 + ... +X_{200}$ of 200 employees is a normal distribution with mean $n \mu$ and standard deviation $\sigma \sqrt{n}$. In this case, sample size $n = 200$, $\mu = E[X_i] = 7000$ and $\sigma(X_i) = 1414$. So, mean of the normal distribution, $n \mu = 7000 \times 200 = 1400000$ and standard deviation of the normal distribution $\sigma \sqrt{n} = 1414 \times \sqrt{200} = 19996.979771956$.

Now to get the probability that the budget 1440000 INR is insufficient for the department employees, we have to find the area under the normal distribution $N(1400000, 19996.979771956)$ in the range $[1440000, +\infty]$ by performing the integration of this normal distribution in this range.

**Question:** In India the mean IQ is 82 and the S.D. of population IQ is 15. If we sample the IQ of 100 people, what is the probability that the average IQ of the sample is not more than 90?

Here, population mean $\mu = 82$, population S.D. $\sigma = 15$, sample size $n = 100$. Now we know, from CLT, that the distribution of the **sample sum** $X^+ = X_1 + X_2 + ... + X_{100}$ is going to be normal $N(n \mu, \sigma \sqrt{n})$. Also it tells us that the distribution of **sample mean** $\bar{X} = \frac{X^+}{n}$ is also going to be normal, with mean $\frac{n \mu}{n} = \mu$ and S.D. $\frac{\sigma \sqrt{n}}{n} = \frac{\sigma}{\sqrt{n}}$.

So, in this case, average IQ of sample mean indicates calculation of mean of distribution of sample mean $\bar{X}$, which is simply $\mu$. Therefore, we can take distribution of sample mean $\bar{X}$ which is the normal distribution $N(\mu, \frac{\sigma}{\sqrt{n}})$.

Here, $\mu = 82$ and $\frac{\sigma}{\sqrt{n}} = \frac{15}{10} = 1.5$.

And finally we have to calculate the area under the normal distribution $N(82, 1.5)$ in the range $[-\infty, 90]$ to find the probability that the average IQ of the sample of 100 people is $\le 90$.   

```python
from scipy.integrate import quad

# Mean and S.D. of normal distribution
mu = 82
sigma = 1.5

norm_dist = lambda x: (1 / (sigma * np.sqrt(2* np.pi))) * np.exp((-(x - mu)**2) / (2 * sigma**2))

print(quad(norm_dist, -100, 90))

# The answer is 1.
```
