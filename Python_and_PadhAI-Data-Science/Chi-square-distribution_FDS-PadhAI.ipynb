## Estimating mean of sample variance $E[S^2]$

### For discrete distribution

For a given sample, the variance $S^2 = \frac{\sum_{i=1}^n (X_i - \bar{X})^2}{n}$,

where, $n$ is the sample size, $X_i$ is the $i^{th}$ element of the sample, $\bar{X}$ is the mean of all elements of that sample. This $S^2$ is known as sample variance.

Now we have to calculate expected value of sample variance $E[S^2]$. What is the relation between $E[S^2]$ to $\mu$ and $\sigma$. What would you like it to be?

We will take the example of dice throw. Consider each sample as three dice throws (i.e., sample size $n = 3$). Sample 1000 such samples. Compute $S^2$. Compute average of the 1000 $S^2$ values.

Remember, for a fair dice, outcomes are 1, 2, 3, 4, 5, 6.

So, mean $ = (1+2+3+4+5+6)/6 = 3.5$ and

varince $ = \frac{(1-3.5)^2 + (2-3.5)^2 + (3-3.5)^2 + (4-3.5)^2 + (5-3.5)^2 + (6-3.5)^2}{6} = 2.916667$.

Note that we are doing two different statistics here. The **average** or **mean** is taken across all samples (here 1000 samples) and **varince** is taken across elements within a sample (here 3 elements of a sample). When we take large number of samples, this average or mean of sample variance $S^2$ of 1000 samples is equal to the expected value of sample varince $E[S^2]$.

Previously we found $E[\bar{X}] = \mu$ i.e., expected value of sample mean is equal to the population mean.

Check here if we get $E[S^2] = \sigma^2$ i.e., if we find expected value of sample varince is equal to the population varince.

```python
import numpy as np
import matplotlib.pyplot as plt

# Testing
# Genrate 3 random numbers between 1 and 6
test_sample = np.random.randint(1, 7, size = 3)
print(test_sample)

# Compute mean and variance of test_sample
print(np.mean(test_sample))
print(np.var(test_sample))


# Final

sample_varince_list = []
for i in range(1000):
    sample = np.random.randint(1, 7, size = 3)
    np.mean(sample)
    np.var(sample)
    sample_varince_list.append(np.var(sample))

sample_varince_array = np.array(sample_varince_list)
print('Size of the varince array or the number of samples = ', np.size(sample_varince_array))
print('The array of all varince valus for all 1000 samples: ', sample_varince_array)
print('The average or mean of 1000 S^2 values of 1000 samples = ', np.mean(sample_varince_array))
```

We get $E[S^2] \approx 2.00$.

```python
plt.hist(sample_varince_array, bins = 20)
plt.show()

# We get discrete histogram plot as the outcomes are discrete in dice rolling, so there are discrete values of sample varince.
```

Clearly, as $n$ increases, $E[S^2]$ increases. And the difference between $E[S^2]$ and $\sigma^2$ is decreasing with increasing $n$. Therefore, one may give a **hypothesis**:

$E[S^2] \rightarrow \sigma^2$ as $n \rightarrow \infty$.

and also, $E[S^2] \propto \mu^0$.

We notice, **$E[S^2]$ is always an underestimate of $\sigma^2$ and become more accurate as the sample size $n$ increases**.

It is not that $E[S^2]$ converges to $\sigma^2$ with an osciallatory behaviour.

According to the definition, $E[S^2] = E[\frac{\sum_{i=1}^n (X_i - \bar{X})^2}{n}]$,

and $\sigma^2 = \frac{\sum_{i=1}^N (X_i - \mu)^2}{N} \approx E[\frac{\sum_{i=1}^n (X_i - \mu)^2}{n}]$.

So, the only difference we have is $\bar{X}$ and $\mu$ i.e, in the case of sample variance, we have sample mean $\bar{X}$ and in the case of population variance, we have population mean $\mu$.

But we know $E[\bar{X}] = \mu$, i.e., $\bar{X}$ can be sometimes $\ge \mu$ or, $\le \mu$. But the expected value of $\bar{X}$ is equal to $\mu$. That is, $\bar{X}$ is an unbiased estimate of $\mu$.

So, even with $E[\bar{X}] = \mu$, why then do we have the underestimation: $E[S^2] < \sigma^2$?

## Estimating variance of sample variance $Var(S^2)$

### Chi square distribution:

In the derivation of $var(S^2)$, we have got:

$\sum_{i=1}^{n} \left(\frac{X_i - \mu}{\sigma} \right)^2 = \frac{(n-1) S_{n-1}^2}{\sigma^2} + \frac{n (\bar{X} - \mu)^2}{\sigma^2}$ .

Here we assume that the RV $X$ follow a normal distribution $N(\mu, \sigma)$ i.e., population values follow a normal distribution.

The term on the LHS is the sum of squres of $n$ I.I.D. random variables (RVs) from a standard normal distribution.

The second term on the the RHS is the square of one RV from a standatrd normal distribution.

Now we want to know about the distribution of $S_{n-1}^2$ from the distribution of sum of squares of $z$-scores i.e., from the distribution of sum of squares of standard normal variables.

**Chi square distribution: Distribution of sum of squares of standard normal variables**

Some people call it "chi squared distribution".

Let $Z_1, Z_2, ..., Z_n$ be independent standard normal variables (we have denoted by $Z$ as the $z$-score transform any normal distribution into a standard normal distribution) i.e., RVs distributed as per standard normal distribution. And we take the sum of squares of them: $Q = \sum_{i=1}^{n} Z_i^2$. $Q$ become a new RV. We are intereseted in studying distribution of $Q$, its mean, S.D., variance etc.

Question: What is distribution of $Q$ as parameterized by $n = $ number of independent standard normal variables under consideration. Let us try plotting for $n = 1, 2, 3, ...$

So, first take $n = 1$. Take 100000 samples from the standard normal distribution. Square all those values. Plot a histogram. Think of the shape of the histogram as an approximation of the distribution of $Q$ for $n = 1$.

Similarly for $n = 2$, take two sets of 100000 samples from the standard normal distribution. Square them independently and add them pair wise, and plot the histogram of the resultant 100000 sums. Think of the shape of the histogram as the distribution of $Q$ for $n = 2$.

Similarly do this for $n = 3, 4, 5,...$   

Our understanding about the distribution of $Q$ is following:

(i) Clearly PDF of $Q$ will take only non-negative values as $Q$ is square of a value or sum of squares of multiple values. PDF of $Q$ will be zero for all negative $Q$ values.

(ii) As the standard normal distribution falls symmetrically and rapidly about zero on both positive and negative $x$-axis, PDF of $Q$ will have almost zero values at large positive $x$-axis value or $Q$ values. PDF of $Q$ will fall as you go away from the origin $Q = 0$.

(iii) How much faster the PDF of $Q$ fall? Does it fall slower or faster than the standard normal distribution? For example, in the PDF, at $Q = 4$, the values come from either from $Z = 2$ or $Z = -2$. Clearly PDF at $Q = 4$ will be larger than it is for standard normal distribution. Also, while for standard normal distribution at $Z = 4$, PDF of $Z$ almost vanishes, for $Q = 4$, PDF of $Q$ will have finite value as the contribtuion comes from $Z = +2$ and $Z = -2$ at this point. i.e., PDF of $Q$ will have longer tail than the PDF of $Z$. PDF of $Q$ will have more area under the tail far away from the origin.

(iv) What happens close to the origin $Q = 0$? Wolud be more or less peaky for the PDF of $Q$ than the standard normal distribution? We know that for any real number $x$, $x^2 < x$ if $x < 1$ meaning if you sqaure the numbers less than 1, they become even smaller and hence move closer to the origin $x = 0$. So, the values in the region $-1 < x < 1$ of the standard normal distribution become even smaller after the square and hence move closer to the origin. So, in the PDF of $Q$, we will have larger density of values closer to the origin $Q = 0$. More peaky behaviour closer to the origin $Q = 0$. Larger density of values increases the probability of $Q$ closer to the origin $Q = 0$.

```python
# For n = 1. Q = Z_1^2.

import numpy as np
import matplotlib.pyplot as plt

Z1 = np.random.standard_normal(100000)

Q = Z1**2

plt.hist(Q)
plt.show()
```

```python
# Joining mid points of histogram by line

y, edges = np.histogram(Q, bins = 100)
centers = 0.5*(edges[1:]+ edges[:-1])
plt.plot(centers, y, '-o')
plt.xlim(0, 20)
plt.show()
```

For $n = 2$, $Q = Z_1^2 + Z_2^2$ where both $Z_1$ and $Z_2$ are distributed as per standard normal distribution. First you take a random sample of $Z_1$ and then take a random sample of $Z_2$, sqaure them separately and add them. This final sum gives one sample of $Q$. Do this for many many samples from $Z_1$ and $Z_2$ to get many samples and PDF of $Q$.

```python
# For n = 2.

import numpy as np
import matplotlib.pyplot as plt

Z1 = np.random.standard_normal(100000)
Z2 = np.random.standard_normal(100000)

Q = Z1**2 + Z2**2

plt.hist(Q)
plt.show()
```

```python
# Joining mid points of histogram by line

y, edges = np.histogram(Q, bins = 100)
centers = 0.5*(edges[1:]+ edges[:-1])
plt.plot(centers, y, '-o')
plt.xlim(0, 30)
plt.show()
```

We see that as $n$ increases, the tail of the distribution of $Q$ increases. That is more area under the tail further away from the origin as $n$ increases. We can think of the distributions for different $n$ values as "transforming area under the curve" becasue all the distributions $Z_1, Z_2, ..., Z_n$ and $Q$ are PDFs having area under the curve is 1. So, we can think as $n$ increases, how the same amout of area under the PDF is getting transferred or squizzed or moved about away from the origin $Q = 0$ i.e., tail of the PDF of $Q$ is getting longer as $n$ increases.

Also we notice that the value close to the origin $Q = 0$ is reducing as $n$ increases (compare the heights of the bar near $Q = 0$ for different $n$ values). We can understand by a small example. Let's take $Q = 0.01$. So what is the probability of $Q$ being $< 0.01$? or $P(Q < 0.01) = ?$ This will be the area of the small strip under the PDF of $Q$ for $Q < 0.01$. It is very clear that for $n = 2$, $Q = Z_1^2 + Z_2^2$ and for $|Z_1| > 0.1$, $Q > 0.01$, independent of the value of $Z_2$. So for $Q < 0.01$, we need **at least** two things: $|Z_1| < 0.1$ **and** $|Z_2| < 0.1$ i.e., $P(Q < 0.01) = P(|Z_1| < 0.1) \times P(|Z_2| < 0.1)$. Notice that, since the $|Z_1|$ and $|Z_2|$ are two independent RVs, we use **multiplication** for both the condition to be satisfied (**and** operation). Though these two conditions are not sufficient, as the **sum** to be $< 0.01$. So we can write, $P(Q < 0.01) < P(|Z_1| < 0.1) \times P(|Z_2| < 0.1)$. So, for $Q$ to be small, both the random numbers $Z_1$ and $Z_2$ have to be small. For $n = 1$, only $Z_1$ has to be small. For larger $n$, all the $Z_1, Z_2, ..., Z_n$ have to be small. Hence, the area about $Z_i = 0$ for all $i$ becomes very very small for large $n$ which reduces the area under the curve for small $Q$ and hence the peak height at $Q = 0$. Basically what happens, as $n$ increases, the density of values close to the origin $Q = 0$ has moved to the right which causes loger tail. i.e., mean or the centre of mass of the PDF of $Q$ shifts to the right as $n$ increases.

So, in general, $P(Q < \epsilon^2) < P(|Z_1| < \epsilon) \times P(|Z_2| < \epsilon) \times ... \times P(|Z_n| < \epsilon)$.

As we discusses above, the area moves to the right as $n$ increases, but where is this area coming from? The area is coming from the small $Q$ value and that is why the area under the PDF decreases for small $Q$ value and this area gets add up in the area under the long tail in the right. ## Distribution of sample variance and Chi square distribution: Two experiments

We have seen that the distribution of scaled version of sample variance $S_{n-1}^2$ (i.e., $\frac{(n-1)S_{n-1}^2}{\sigma^2}$) is same as the chi square distribution with $(n-1)$ degrees of freedom i.e., $\chi^2(n-1)$, where $n$ is the sample size.

Now we will compare the distribution of sample variance and chi square distribution for two experiments which we did earlier, one is for continious distribution (standard normal distribution) and another is for discrete distribution (uniform distribution for dice rolling case).

### For continious distribution

Recall our earlier example of radom sampling from standard normal distribution. Basically we will plot the two earlier plots of (i) histogram of $\frac{(n-1)S_{n-1}^2}{\sigma^2}$ with bin size = 1 and (ii) the chi sqaure distribution $\chi^2(n-1)$, in a single plot. We did both the plots earlier but separately. But now we will show them in a single plot.

Since here the population has a normal distrbution, we can compute both the mean and variance of the sample variance i.e., $E[S_{n-1}^2]$ and $Var(S_{n-1}^2)$.

Always remember that if the sample size is $n$, then the distribution of the sample variance (actually scaled version of the sample variance: $\frac{(n-1)S_{n-1}^2}{\sigma^2}$) is the chi square distribution with $(n-1)$ d.o.f. i.e., $\chi^2(n-1)$.

```python
# For sample size n = 2.

import numpy as np
import matplotlib.pyplot as plt

# (i) For distribution of scaled version of sample variance: (n-1)*S_{n-1}^2 / \sigma^2.

# Random sample from normal distribution. Here mean = 0 and S.D. = 1 = Variance. Hence it is a standard normal distribution.
# print(np.random.normal(0, 1, size = 3))

# Sample size
n = 2

# Population variance
sigma = 1

sample_variance_list = []
for i in range(100000):
    sample = np.random.normal(0, 1, size = n)
    np.var(sample, ddof = 1)   # ddof = “Delta Degrees of Freedom”: the divisor used in the calculation is (n - ddof), where n represents the number of elements or sample size in this case. By default ddof is zero. To get the unbiased estimte of the sample variance, we divide by (n-1).
    sample_variance_list.append(np.var(sample))

sample_varince_array = np.array(sample_variance_list)
print('Number of samples = ', np.size(sample_varince_array))
# print('The array of sample variance values of all the samples: ', sample_varince_array)
print('Mean or average of sample variance across all the samples = ', np.mean(sample_varince_array))


# (ii) For Chi square distribution with (n-1) degrees of freedom. Since n = 2 above, (n-1) = 1 for the chi square distribution
# below.

import numpy as np
import matplotlib.pyplot as plt

Z1 = np.random.standard_normal(100000)

Q = Z1**2


# Plot

# Histogram plot of scaled version of sample variance.
plt.hist(((n-1)*sample_varince_array)/sigma**2, bins = 100, density = True)   # density = True makes the area under the distriution = 1.
plt.xlabel('Sample variance $S^2$')
plt.ylabel('Frequency')

# Line plot of Chi square distribution with (n-1) d.o.f.
y, edges = np.histogram(Q, bins = 100, density = True)   # density = True makes the area under the distriution = 1.
centers = 0.5*(edges[1:]+ edges[:-1])
plt.plot(centers, y, '-o')
plt.xlim(0, 6)

plt.title('Comparision of distribution of $(n-1)S_{n-1}^2 / \sigma^2$ and $\chi^2(n-1)$ for $n = 2$')
plt.legend(['$\chi^2(n-1)$', '$(n-1)S_{n-1}^2 / \sigma^2$'])

plt.show()
```

For comparison of histogram and PDF, we convert the histogram into probability.

We see a good approximation between the distribution of scaled version of the sample variance and chi square distribution, but not perfectly matching.

### For discrete distribution

Let's take the example of dice rolling. Consider each sample as $n$ dice throwing or rolling. So, each sample consists of $n$ elements. Consider such 1000 samples. Compute the sample variance $S^2$. Compute distribution of 1000 $S^2$ values. Notice that, population has only 6 outcomes and all of them are equaly likely. Thus this is a discrete uniform (not normal!!) distribution. Hence our formula of computing variance of sample variance does not hold good, and also the distribution of sample variance does not have chi square distribution. However, we will compare the actual distribution with the corresponding chi square distribution.

```python
# For sample size n = 2.

import numpy as np
import matplotlib.pyplot as plt

# (i) For distribution of scaled version of sample variance: (n-1)*S_{n-1}^2 / \sigma^2.

# Random sample from normal distribution. Here mean = 0 and S.D. = 1 = Variance. Hence it is a standard normal distribution.
# print(np.random.normal(0, 1, size = 3))

# Sample size
n = 2

# Population variance
sigma = 1

sample_variance_list = []
for i in range(1000):
    sample = np.random.randint(1, 7, size = n)   # Random sample of n integer elements among the outcomes: 1, 2, ..., 6.
    np.var(sample, ddof = 1)   # ddof = “Delta Degrees of Freedom”: the divisor used in the calculation is (n - ddof), where n represents the number of elements or sample size in this case. By default ddof is zero. To get the unbiased estimte of the sample variance, we divide by (n-1).
    sample_variance_list.append(np.var(sample))

sample_varince_array = np.array(sample_variance_list)
print('Number of samples = ', np.size(sample_varince_array))
# print('The array of sample variance values of all the samples: ', sample_varince_array)
print('Mean or average of sample variance across all the samples = ', np.mean(sample_varince_array))


# (ii) For Chi square distribution with (n-1) degrees of freedom. Since n = 2 above, (n-1) = 1 for the chi square distribution
# below.

import numpy as np
import matplotlib.pyplot as plt

Z1 = np.random.standard_normal(1000)

Q = Z1**2


# Plot

# Histogram plot of scaled version of sample variance.
plt.hist(((n-1)*sample_varince_array)/sigma**2, bins = 10, density = True)   # density = True makes the area under the distriution = 1.
plt.xlabel('Sample variance $S^2$')
plt.ylabel('Frequency')

# Line plot of Chi square distribution with (n-1) d.o.f.
y, edges = np.histogram(Q, bins = 10, density = True)   # density = True makes the area under the distriution = 1.
centers = 0.5*(edges[1:]+ edges[:-1])
plt.plot(centers, y, '-o')
plt.xlim(0, 6)

plt.title('Comparision of distribution of $(n-1)S_{n-1}^2 / \sigma^2$ and $\chi^2(n-1)$ for $n = 2$')
plt.legend(['$\chi^2(n-1)$', '$(n-1)S_{n-1}^2 / \sigma^2$'])

plt.show()
```

When the sample size is very large, say $n = 100$, the distribution of sample variance actually become a normal distribution (by central limit theorem).

```python
# For sample size n = 100.

import numpy as np
import matplotlib.pyplot as plt

# (i) For distribution of scaled version of sample variance: (n-1)*S_{n-1}^2 / \sigma^2.

# Random sample from normal distribution. Here mean = 0 and S.D. = 1 = Variance. Hence it is a standard normal distribution.
# print(np.random.normal(0, 1, size = 3))

# Sample size
n = 100

# Population variance
sigma = 1

sample_variance_list = []
for i in range(1000):
    sample = np.random.randint(1, 7, size = n)   # Random sample of n integer elements among the outcomes: 1, 2, ..., 6.
    np.var(sample, ddof = 1)   # ddof = “Delta Degrees of Freedom”: the divisor used in the calculation is (n - ddof), where n represents the number of elements or sample size in this case. By default ddof is zero. To get the unbiased estimte of the sample variance, we divide by (n-1).
    sample_variance_list.append(np.var(sample))

sample_varince_array = np.array(sample_variance_list)
print('Number of samples = ', np.size(sample_varince_array))
# print('The array of sample variance values of all the samples: ', sample_varince_array)
print('Mean or average of sample variance across all the samples = ', np.mean(sample_varince_array))


# (ii) For Chi square distribution with (n-1) degrees of freedom. Since n = 100 above, (n-1) = 99 for the chi square distribution
# below.

sum = np.zeros(1000)

for i in range(100):
    Z = np.random.standard_normal(1000)
    # print(Z)
    # print(Z**2)

    sum = sum + Z**2

Q = sum


# Plot

# Histogram plot of scaled version of sample variance.
plt.hist(((n-1)*sample_varince_array)/sigma**2, bins = 10, density = True)   # density = True makes the area under the distriution = 1.
plt.xlabel('Sample variance $S^2$')
plt.ylabel('Frequency')

# Line plot of Chi square distribution with (n-1) d.o.f.
y, edges = np.histogram(Q, bins = 10, density = True)   # density = True makes the area under the distriution = 1.
centers = 0.5*(edges[1:]+ edges[:-1])
plt.plot(centers, y, '-o')

plt.title('Comparision of distribution of $(n-1)S_{n-1}^2 / \sigma^2$ and $\chi^2(n-1)$ for $n = 100$')
plt.legend(['$\chi^2(n-1)$', '$(n-1)S_{n-1}^2 / \sigma^2$'])

plt.show()
```
