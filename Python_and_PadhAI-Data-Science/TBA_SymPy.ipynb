# Lieb lattice

```python
# Band structure of Lieb lattice along Gamma-M path.
# When the onsite energies are same: e_s = e_p.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1   # Lattice constant of the Lieb lattice.
t = 1   # The NN interation strength.
e_s = 2   # On-site energy for s orbital.
e_p = 2   # On-site energy for p orbital.

# The array of k (Note that for Gamma-M path, k_x = k_y = k):
k = np.linspace(0, 6.28, 1000)   # We are taking this array upto 2*pi/a with a = 1.

# Energy eigenvalues along Gamma-M path (k_x = k_y = k)
E_1 = np.full(1000, e_s)    # Array of shape (1000, 1) having same value e_s.
E_2 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 32 * t**2 * np.cos(0.5*k*a)**2))
E_3 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 32 * t**2 * np.cos(0.5*k*a)**2))

# Plot
plt.plot(k, E_1)
plt.plot(k, E_2)
plt.plot(k, E_3)
plt.legend(['E_1', 'E_2', 'E_3'])
plt.title('TBA band dispersion of Lieb lattice for NN interaction')
plt.xlabel('k (along $\Gamma - M$)')
plt.ylabel('Energy')
plt.show()
```

When the onsite energies are same, the three eigenvalues are degenerate at the $M$ point. Along the $\Gamma - M$ path, flat band passes through the Dirac cone.

```python
# Band structure of Lieb lattice along Gamma-M path.
# When the onsite energies are not same : e_s /= e_p.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1   # Lattice constant of the Lieb lattice.
t = 1   # The NN interation strength.
e_s = 1.5   # On-site energy for s orbital.
e_p = 2   # On-site energy for p orbital.

# The array of k (Note that for Gamma-M path, k_x = k_y = k):
k = np.linspace(0, 6.28, 1000)   # We are taking this array upto 2*pi/a with a = 1.

# Energy eigenvalues along Gamma-M path (k_x = k_y = k)
E_1 = np.full(1000, e_s)   # Array of shape (1000, 1) having same value e_s.
E_2 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 32 * t**2 * np.cos(0.5*k*a)**2))
E_3 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 32 * t**2 * np.cos(0.5*k*a)**2))

# Plot
plt.plot(k, E_1)
plt.plot(k, E_2)
plt.plot(k, E_3)
plt.legend(['E_1', 'E_2', 'E_3'])
plt.title('TBA band dispersion of Lieb lattice for NN interaction')
plt.xlabel('k (along $\Gamma - M$)')
plt.ylabel('Energy')
plt.show()
```

When the onsite energies are not same, along the $\Gamma - M$ path, gap opens up at the Dirac cone.

# Consider Ag-5$s$, Sn-5$p_x$, Sn-5$p_y$ orbitals per unit cell in $x = 0.25, p(2\times 2)$ cell of Sn/Ag(001)

In this case we consider conly nearest neighbour (NN) interactions and we consider Ag-5$s$, Sn-5$p_x$ and Sn-5$p_y$ orbitals per unit cell. So, we have done the calculations of TBA which gives the $(4 \times 4)$ Hamiltonian matrix $H(\tilde {\textbf{k}})$ in the momentum space. We have solved for the eigenvalues and we will plot here the band dispersion relations for different conditions and paths.

```python
# The lattice constant of the p(2x2) cell will be 2a and hence the primitive translation vectors in reciprocal lattice will be
# b1 = b2 = 2*pi / 2a = pi/a. So, the Brillouin zone boundary is at k_x = pi/2a on the k_x axis and at k_y = pi/2a on the k_y axis.
# As we took, a = 1, so the BZ boundary will be at k_x = pi/2 = 1.57 on the k_x axis and same on the k_y axis.

# Here we plot with k_y = 0 i.e., along the Gamma-X line.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1   # The surface lattice constant of the (1x1) cell of clean Ag(001).
V = 1   # The s-p_x and s-p_y interation strengths.
e_s = 2   # On-site energy for s orbital.
e_p = 2.5   # On-site energy for p orbital.

# The array of k_x:
k_x = np.linspace(0, 3.14, 1000)   # We are taking this array upto pi/a with a = 1.

# Energy eigenvalues for k_y = 0:
E_1 = np.full(1000, (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2)))
E_2 = np.full(1000, (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2)))
E_3 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 8 * V**2 * (1 + np.cos(2*k_x*a))))
E_4 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 8 * V**2 * (1 + np.cos(2*k_x*a))))


# Plot
plt.plot(k_x, E_1)
plt.plot(k_x, E_2)
plt.plot(k_x, E_3)
plt.plot(k_x, E_4)
plt.legend(['E_1', 'E_2', 'E_3', 'E_4'])
plt.title('TBA band dispersion by taking Ag-5$s$, Sn-5$p_x$, Sn-5$p_y$ orbitals NN interactions')
plt.xlabel('k_x (along $\Gamma - X$)')
plt.ylabel('Energy')
plt.show()
```

```python
# Onsite energies are same: e_s = e_p.

# The lattice constant of the p(2x2) cell will be 2a and hence the primitive translation vectors in reciprocal lattice will be
# b1 = b2 = 2*pi / 2a = pi/a. So, the Brillouin zone boundary is at k_x = pi/2a on the k_x axis and at k_y = pi/2a on the k_y axis.
# As we took, a = 1, so the BZ boundary will be at k_x = pi/2 = 1.57 on the k_x axis and same on the k_y axis.

# Here we plot with k_x = k_y = k i.e., along the Gamma-M line.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1   # The surface lattice constant of the (1x1) cell of clean Ag(001).
V = 1   # The s-p_x and s-p_y interation strengths.
e_s = 2   # On-site energy for s orbital.
e_p = 2   # On-site energy for p orbital.

# The array of k_x:
k_x = np.linspace(0, 3.14, 1000)   # We are taking this array upto pi/a with a = 1.

# Energy eigenvalues for k_x = k_y = k:
E_1 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_2 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_3 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_4 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))


# Plot
plt.plot(k_x, E_1)
plt.plot(k_x, E_2)
plt.plot(k_x, E_3)
plt.plot(k_x, E_4)
plt.legend(['E_1', 'E_2', 'E_3', 'E_4'])
plt.title('TBA band dispersion of 0.25 ML $p(2 x 2)$ cell by taking Ag-5$s$, Sn-5$p_x$, Sn-5$p_y$ orbitals considering NN interactions')
plt.xlabel('k (along $\Gamma - M$)')
plt.ylabel('Energy')
plt.show()
```

```python
# Onsite energies are not same: e_s /= e_p.

# The lattice constant of the p(2x2) cell will be 2a and hence the primitive translation vectors in reciprocal lattice will be
# b1 = b2 = 2*pi / 2a = pi/a. So, the Brillouin zone boundary is at k_x = pi/2a on the k_x axis and at k_y = pi/2a on the k_y axis.
# As we took, a = 1, so the BZ boundary will be at k_x = pi/2 = 1.57 on the k_x axis and same on the k_y axis.

# Here we plot with k_x = k_y = k i.e., along the Gamma-M line.

import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1   # The surface lattice constant of the (1x1) cell of clean Ag(001).
V = 1   # The s-p_x and s-p_y interation strengths.
e_s = 2   # On-site energy for s orbital.
e_p = 2.5   # On-site energy for p orbital.

# The array of k_x:
k_x = np.linspace(0, 3.14, 1000)   # We are taking this array upto pi/a with a = 1.

# Energy eigenvalues for k_x = k_y = k:
E_1 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_2 = (0.5 * (e_s + e_p)) + (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_3 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))
E_4 = (0.5 * (e_s + e_p)) - (0.5 * np.sqrt((e_s - e_p)**2 + 16 * V**2 * (np.cos(k_x*a))**2))


# Plot
plt.plot(k_x, E_1)
plt.plot(k_x, E_2)
plt.plot(k_x, E_3)
plt.plot(k_x, E_4)
plt.legend(['E_1', 'E_2', 'E_3', 'E_4'])
plt.title('TBA band dispersion of 0.25 ML $p(2 x 2)$ cell by taking Ag-5$s$, Sn-5$p_x$, Sn-5$p_y$ orbitals considering NN interactions')
plt.xlabel('k (along $\Gamma - M$)')
plt.ylabel('Energy')
plt.show()
```

We notice that when $e_s = e_p$ i.e., onsite energies of $s$- and $p$-orbitals are same, two doubly degenerate bands touch each other at the $\bar{M}$ point of the Brillouin zone of $p(2\times 2)$ cell (the SCBZ). Thus Dirac cone and linear dispersion is observed. And when $e_s \ne e_p$, the gap opens up at the $\bar{M}$ point.
