```python
def f(x):
    n = len(x)
    idc = (n // 2, n // 2) if n%2==1 else ((n // 2) - 1, (n // 2))
    return (x[idc[0]] + x[idc[1]]) / 2

x = [4, 8, 16, 32]
print(f(x))
```

```python
def f(x):
    l = {}
    for v in x:
        if v in l:
            l[v] += 1
        else:
            l[v] = 0
    return list(l.keys())[list(l.values()).index(max(l.values()))]

x = [1, 1, 1, 2, 2, 2, 2, 2, 3, 3]
print(f(x))
```

```python
def f(a, b):
    at = list(a)
    for i, v in enumerate(a):
        if v in b:
            at.pop(i)
    return set(at)

a = {2, 4, 6, 8, 10}
b = {1, 4, 9}

print(f(a, b))
```

```python
def f(x, a):
    c = 0
    for v in x:
        c += 1 if a==v else 0
    return c

x = ['Kachha', 'papad', 'pakka', 'papad']
print(f(x, x[1]))
```
